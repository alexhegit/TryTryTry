```python
import cv2
import time, timeit
from timeit import default_timer as timer
```

```python
from matplotlib.pyplot import imshow
%matplotlib inline
```

```python
IMG = './292.bmp'
```

```python
img = cv2.imread(IMG)
print('Image Dimensions :', img.shape)
imshow(img)
```

```python
g_img = cv2.imread(IMG, cv2.IMREAD_GRAYSCALE)
```

```python
print('Image Dimensions :', g_img.shape)
```

```python
imshow(img, cmap='gray')
imshow(img)
```

```python
m_img = cv2.morphologyEx(g_img, cv2.MORPH_OPEN, (5,5))
```

```python
imshow(m_img, cmap='gray')
```

```python
t1 = timer()
ttt = cv2.morphologyEx(g_img, cv2.MORPH_OPEN, (5,5))
t2 = timer()
print("Time Used", t2 - t1, "second")
```

```python
t1 = time.time()
ttt = cv2.morphologyEx(g_img, cv2.MORPH_OPEN, (5,5))
t2 = time.time()
print("Time Used", t2 - t1, "second")
```
